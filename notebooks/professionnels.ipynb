# Tous les professionnels sont-ils égaux ?

On s'intéresse ici aux Artisans, commerçants et chefs d'entreprise (Catégorie Socio-Professionnelle CS2 dans la nomenclature de l'Insee).
Ces derniers peuvent avoir besoin des services des structures DGFIP dédiées aux professionnelles.

```python
from dgfip.distance import mean_distances
```

```python
pro = mean_distances(categories=['C17_POP15P_CS2'], public='professionnels')
```

```python
from dgfip.read import get_dep
dep = get_dep()
data = dep.merge(pro, left_on='code', right_on='DEP')
```

```python
data[data.service=='Centre de Finances publiques'].plot('distance_avg', legend=True, scheme='Quantiles')
```

On observe une forte disparité géographique avec les zones denses - l'Île de France, le Nord, le Rhône et les Bouches du Rhônes - qui ont une distance moyenne faible tandis que les zones rurales sont moins bien pourvues.
